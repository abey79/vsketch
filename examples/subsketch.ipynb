# Sub-sketch

The `sketch()` method draws the content of a sub-sketch in the main sketch, preserving formatting (ie. layers) but applying the current transformation matrix. This makes it easy to create reusable drawing elements

```python
%load_ext autoreload
%autoreload 2

#  use `widget` for interactive plot
%matplotlib inline


from matplotlib import rcParams

rcParams["figure.figsize"] = (14, 10)
```

```python
import vsketch

# create a stick figurestick figure
sub = vsketch.Vsketch()
sub.detail(0.01)
sub.rect(0, 0, 1, 2)
sub.circle(0.5, -0.5, 1)
sub.line(0, 0, -0.5, 1)
sub.line(1, 0, 1.5, 1)
sub.line(0, 2, -0.3, 4)
sub.line(1, 2, 1.3, 4)

# use the stick figure
vsk = vsketch.Vsketch()
vsk.size("a4", landscape=True)
vsk.scale("1cm")


for i in range(8):
    with vsk.pushMatrix():
        vsk.scale(0.95 ** i)
        vsk.rotate(8 * i, degrees=True)
        vsk.sketch(sub)
    
    vsk.translate(3, 0)


vsk.plot()
vsk.save("subsketch_basic.svg")
```
