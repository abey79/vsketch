```python
import vsketch

vsk = vsketch.Vsketch()
vsk.size("a4", landscape=True)
vsk.scale("1cm")

# the default is no fill and stroke to layer 1
vsk.square(0, 0, 4)
vsk.circle(2, 9, 4)

vsk.translate(5, 0)

# add some fill to layer 2
vsk.fill(2)
vsk.penWidth("1mm", 2)
vsk.square(0, 0, 4)
vsk.circle(2, 9, 4)

vsk.translate(5, 0)

# remove stroke and set fill to layer 3 with a thicker pen
vsk.fill(3)
vsk.penWidth("3mm", 3)
vsk.noStroke()
vsk.square(0, 0, 4)
vsk.circle(2, 9, 4)


vsk.plot()
vsk.save("fill.svg")
```
